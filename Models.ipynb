```python
from tqdm.notebook import tqdm
from catboost import CatBoostClassifier
import sklearn
from sklearn.model_selection import train_test_split
from datetime import datetime
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
import pandas as pd
```

## Модель на признаках пользователя и контента

```python
non_graph = pd.read_csv('vlakhov_shalava_features.csv')
```

```python
del non_graph['Unnamed: 0']
```

```python
non_graph = non_graph.rename(columns={"taget": "target"})
```

```python
columns = list(non_graph.columns)
```

```python
l = ['user_id', 'subscriptions', 'posts', 'texts', 'embeddings_texts', 'repost', 'comments', 'likes', 'reposts_count', 'target', 'subscriptions_count']
feature_columns = [col for col in columns if col not in l]
```

```python
reposts = list(non_graph.repost)
```

```python
import ast
```

```python
reposts_new = []
for r in reposts:
    if type(r) != float and type(r) != int:
        new = ast.literal_eval(r)
        reposts_new.append(new)
    elif type(r)== float:
        reposts_new.append(new)
    else:
        reposts_new.append(None)
```

```python
reposts_mean = []
for r in reposts_new:
    if type(r) == list:
        new = sum(r)
        reposts_mean.append(new)
    elif type(r) == None:
        reposts_mean.append(None)
    else:
        reposts_mean.append(r)
```

```python
non_graph['repost_bull_count'] = reposts_mean
```

```python
cities = list(non_graph.city)
```

```python
cities_new = []
for city in cities:
    if type(city) == str:
        cities_new.append(city)
    else:
        cities_new.append('NaN')
```

```python
countries = list(non_graph.country)
countries_new = []
for country in countries:
    if type(country) == str:
        countries_new.append(country)
    else:
        countries_new.append('NaN')
```

```python
non_graph['city'] = cities_new
```

```python
non_graph['country'] = countries_new
```

```python
X_train, X_test, y_train, y_test = train_test_split(non_graph[feature_columns], non_graph['target'], test_size=0.33, random_state=42)
```

```python
cat_features = [1, 2]
model = CatBoostClassifier(cat_features=cat_features,
                           random_seed=3)
start_time = datetime.now()
model.fit(X_train, y_train)
target_names=['реальный', 'бот']
```

```python
report = classification_report(y_test, model.predict(X_test),
                              target_names=target_names)
print(report)
print('Затраченное время:', datetime.now() - start_time)
```

```python
def importances(model, feature_columns):
    fuck = sorted(list(zip(list(model.feature_importances_), feature_columns)), reverse=True)
    fuck2 = []
    counter = 0
    for f in fuck:
        if len(f[1]) <= 3:
            counter += f[0]
        else:
            fuck2.append(f)
    if counter != 0:
        fuck2.append((counter, 'embeddings'))
    print(sorted(fuck2, reverse=True))
```

```python
importances(model, feature_columns)
```

## Модель на признаках графов

```python
graph_metrics = pd.read_csv('graph_metrics_all.csv', index_col=False)
```

```python
del graph_metrics['Unnamed: 0']
```

```python
X_train, X_test, y_train, y_test = train_test_split(graph_metrics[['max_dists', 'mean_dists', 'degrees', 'closenesses_real']], 
                                                    graph_metrics['target'], test_size=0.33, random_state=42)
```

```python
model_graph = CatBoostClassifier(random_seed=3)
start_time = datetime.now()
model_graph.fit(X_train, y_train)
target_names=['бот', 'реальный']
report = classification_report(y_test, model_graph.predict(X_test),
                              target_names=target_names)
print(report)
print('Затраченное время:', datetime.now() - start_time)
```

```python
importances(model_graph, ['max_dists', 'mean_dists', 'degrees', 'closenesses_real'])
```
